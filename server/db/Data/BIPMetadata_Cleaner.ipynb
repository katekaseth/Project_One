```python
import pandas as pd
```

```python
data = pd.read_csv("BIPMetadata.csv")
created = data["created"]
updated = data["updated"]
```

```python
def format_date(date_column):
    new_dates = []
    for date in date_column:
        month = date[0:date.find('/')]
        date = date[date.find('/')+1:]
        day = date[0:date.find('/')]
        year = date[date.find('/')+1:]

        if (len(month) == 1):
            month = "0" + month
        if (len(day) == 1):
            day = "0" + day
        newDate = year + "-" + month + "-" + day
        new_dates.append(newDate)
    return new_dates
```

```python
new_created = format_date(created)
new_updated = format_date(updated)
```

```python
new_query = []
for query in data["sql_query"]:
    if (len(query) > 5000):
        query = query[0:5000]
    new_query.append(query)
```

```python
new_d = []
for d in data.description:
    if (len(d) > 200):
        d = d[0:200]
    new_d.append(d)
```

```python
data["updated"] = new_updated
data["created"] = new_created
data["sql_query"] = new_query
data["description"] = new_d
data
```

```python
data.to_csv("BIPMetadata_cleaned.csv") 
```
